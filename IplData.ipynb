```python
mprice=df["ValueinCR"].mean()
meprice=df["ValueinCR"].median()
sns.distplot(df["ValueinCR"])
plt.axvline(mprice,color="g")
plt.axvline(meprice,color="b")
plt.grid(True)
plt.show()
```

```python
#count plot
sns.countplot(data=df,x="Team")
plt.grid(True)
plt.show()
```

```python
sns.countplot(data=df,x="Team",hue="CatchesTaken")
plt.grid(True)
plt.show()
```

```python
#sub plot
plt.figure(1)
plt.subplot(2,2,4)
df["Batting Style"].value_counts().plot(figsize=(6,6),kind="bar",color="purple")
plt.title("Types Of Batting Style")
plt.xlabel("Batting Stance")
plt.ylabel("NO of players")

plt.subplot(2,2,3)
df["100s"].value_counts().plot(figsize=(6,6),kind="bar",color="red")
plt.title("100sof Players ")
plt.xlabel("century")
plt.ylabel("Difference")

plt.subplot(2,2,2)
df["Team"].value_counts().plot(figsize=(6,6),kind="bar",color="green")
plt.title("T20 Teams")
plt.xlabel("Names of team")
plt.ylabel("scale")

plt.subplot(2,2,1)
df["National Side"].value_counts().plot(figsize=(6,6),kind="bar",color="blue")
plt.title("Nation of Players team")
plt.xlabel("Names of team ")
plt.ylabel("scale ")

plt.tight_layout()
plt.show()
```
